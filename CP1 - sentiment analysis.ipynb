```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import re
from transformers import pipeline
```

```python
ikea_reviews = pd.read_csv('ikea_reviews_language.csv')[:500]
```

```python
# Filter for English reviews
english_reviews = ikea_reviews[ikea_reviews['language'] == 'English']
```

```python
sentiment_analysis = pipeline("sentiment-analysis",model="siebert/sentiment-roberta-large-english")
```

```python
#### Read temp results, or create one if there is no result file yet
try:
    temp_sentiment_eng = pd.read_csv("temp_sentiment_eng.csv")
except:
    temp_sentiment_eng = pd.DataFrame(columns=['label', 'score'])

#### Run the sentiment analysis model on each review one at a time.
try:
    count = 0
    for eac in english_reviews['text'].loc[len(temp_sentiment_eng):]:
        try:
            eac_res = sentiment_analysis(eac)  # Perform sentiment analysis
        except:
            print("Sentence too long, taking the first 512 characters")
            eac_res = sentiment_analysis(eac[:512])
        
        temp_sentiment_eng = pd.concat([temp_sentiment_eng, pd.DataFrame(eac_res)])

        #### Save every 100 reviews
        if count % 100 == 0:
            print(count)
            temp_sentiment_eng.to_csv("temp_sentiment_eng.csv")

        count += 1

#### Save if an error occurs
except:
    print("Hit an error. Stopped.")
    temp_sentiment_eng.to_csv("temp_sentiment_eng.csv")
```

```python
sentiment_results_eng = pd.read_csv('temp_sentiment_eng.csv')
```

```python
text_df=pd.DataFrame(sentiment_results_eng)
```

```python
english_reviews['sentiment'] = text_df['label']
english_reviews['sentimentvalue'] = text_df['score']
```

```python
ikea_reviews['sentiment'] = text_df['label']
ikea_reviews['sentimentvalue'] = text_df['score']
```
